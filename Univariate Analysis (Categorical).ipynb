```python
ball = pd.read_excel("D:\Manipal Pro learn\Term1\EDA\Dataset\IPL Data\Ball_by_Ball.xlsx")
match = pd.read_excel("D:\Manipal Pro learn\Term1\EDA\Dataset\IPL Data\Match.xlsx")
player = pd.read_excel("D:\Manipal Pro learn\Term1\EDA\Dataset\IPL Data\Player.xlsx")
season = pd.read_excel("D:\Manipal Pro learn\Term1\EDA\Dataset\IPL Data\Season.xlsx")
team = pd.read_excel("D:\Manipal Pro learn\Term1\EDA\Dataset\IPL Data\Team.xlsx")
player_match = pd.read_excel("D:\Manipal Pro learn\Term1\EDA\Dataset\IPL Data\Player_Match.xlsx")
```

```python
#crosstab
Won_match = pd.crosstab(index=match['Match_Winner_Id'],columns='count').reset_index()
Won_match.head()
```

```python
Team_wise_win = pd.merge(Won_match,team[['Team_Id','Team_Short_Code']],left_on="Match_Winner_Id",right_on='Team_Id')
Team_wise_win[['Team_Short_Code','count']].sort_values(by='count',ascending = False)
#Team_wise_win[['Team_Short_Code','count']]
Team_wise_win = Team_wise_win[['Team_Short_Code','count']]
Team_wise_win.head()
```

```python
# percentage win

team_home_played = pd.crosstab(index=match['Team_Name_Id'],columns='Matches Played').reset_index()
#print(team_home_played.head())
team_away_played = pd.crosstab(index=match['Opponent_Team_Id'],columns='Matches Played').reset_index()
#print(team_away_played.head())
total_match = team_home_played['Matches Played'] + team_away_played['Matches Played']
Team_wise_win.insert(2,'Total',list(total_match))
Team_wise_win
```

```python
print(Team_wise_win[['Team_Short_Code','count','Total']])
Team_wise_win['% win'] = Team_wise_win['count'] / Team_wise_win['Total']
```

```python
percent_win = Team_wise_win[['Team_Short_Code','count','Total','% win']].sort_values(by='% win',ascending = False)
percent_win
```

```python
#season title winners
match_grp_season = match.groupby('Season_Id')
season_list = list()
team_list = list()
for season,group in match_grp_season:
    sort = group.sort_values(by='Match_Date',ascending=False)
    season_list.append(season)
    team_list.append(int(sort.iloc[0,13]))
    
season_winner = pd.DataFrame(team_list,index = season_list,columns=['Team_Id'])
season_win_count = pd.crosstab(index=season_winner['Team_Id'],columns='win count')
title_winner = pd.merge(team,season_win_count,how='left',left_on='Team_Id',right_on='Team_Id')
title_winner = title_winner.fillna(0)
title_winner[['Team_Short_Code','win count']]
```

```python
#Two way cross table 
#no. of matches won by each team while batting first and secod

match_won_type = pd.crosstab(index=match['Match_Winner_Id'],columns=match['Win_Type']).reset_index()
match_won_type
```

```python
Team_wise_win.insert(3,'tie',match_won_type['Tie'])
```

```python
Team_wise_win.insert(4,'Batting First',match_won_type['by runs'])
Team_wise_win.insert(5,'Batting Second',match_won_type['by wickets'])
Team_wise_win
```

```python
Team_wise_win['% Batting 1st win'] = Team_wise_win['Batting First'] / Team_wise_win['count']
Team_wise_win['% Batting 2nd win'] = Team_wise_win['Batting Second'] / Team_wise_win['count']
Team_wise_win
```

```python
#alternate
Team_wise_win = match.groupby(['Match_Winner_Id','Win_Type']).size().reset_index()
Final_team_wise_win = Team_wise_win.pivot('Match_Winner_Id','Win_Type',0)
Final_team_wise_win = Final_team_wise_win.fillna(0)
Final_team_wise_win['Total Matches won'] = Final_team_wise_win['Tie'] + Final_team_wise_win['by runs'] + Final_team_wise_win['by wickets']
Final_team_wise_win
```

```python
#Top5 high scoring batsmen
striker_list = list()
run_scored= list()
for id,group in ball.groupby('Striker_Id'):
    striker_list.append(id)
    run_scored.append(group['Batsman_Scored'].sum())
striker_runs = pd.DataFrame(run_scored,striker_list).reset_index()
striker_runs.columns = ['Batsman_Id','Runs Scored']
player_run_scored = pd.merge(player,striker_runs,how='left',left_on='Player_Id',right_on='Batsman_Id')
top_5_batsmen = player_run_scored[['Player_Name','Runs Scored']].sort_values(by='Runs Scored',ascending=False).head()
top_5_batsmen
```

```python
#Top5 highest wicket taking bowlers
bowler_list = list()
wickets_taken= list()
for id,group in ball[ball['Dissimal_Type'].isin(['bowled','caught','caught and bowled','lbw'])].groupby('Bowler_Id'):
        bowler_list.append(id)
        wickets_taken.append(group['Dissimal_Type'].count())
bowler_wicket = pd.DataFrame(wickets_taken,bowler_list).reset_index()
bowler_wicket.columns = ['Bowler_Id','Wickets taken']
bowler_wicket_taken = pd.merge(player,bowler_wicket,how='left',left_on='Player_Id',right_on='Bowler_Id')
bowler_wicket_taken = bowler_wicket_taken[['Player_Name','Wickets taken']].dropna()
top_5_bowlers = bowler_wicket_taken.sort_values(by='Wickets taken',ascending=False).head()
top_5_bowlers
```

# Multi dimensional table

```python
#Team with highest average match aggregate(average score)

#Total matches
team_home_played = pd.crosstab(index=match['Team_Name_Id'],columns='Matches Played').reset_index()
team_away_played = pd.crosstab(index=match['Opponent_Team_Id'],columns='Matches Played').reset_index()
total_match = team_home_played['Matches Played'] + team_away_played['Matches Played']
total_match = pd.DataFrame(total_match)
teamscore = ball.groupby(['Team_Batting_Id'])['Batsman_Scored'].sum().reset_index()
teamscore.insert(2,'Matches',total_match['Matches Played'])
teamscore['Average'] = teamscore['Batsman_Scored'] / teamscore['Matches']
teamscore[teamscore['Average'] == max(teamscore['Average'])]
teamscore
```

```python
#Team scoring maximum runs in each season
overall_team_score = ball[['Season_Id','Team_Batting_Id','Batsman_Scored']].groupby(['Season_Id','Team_Batting_Id']).sum().reset_index()
Season_top_team = pd.DataFrame()
for name,group in overall_team_score.groupby('Season_Id'):
    Season_top_team = Season_top_team.append(group[group['Batsman_Scored'] == max(group['Batsman_Scored'])])
Season_top_team
```

# Univariate Analysis (Categorical data)

```python
#Does batting first help to score more runs
Innings_score = ball.groupby(['Innings_Id'])['Batsman_Scored'].sum().reset_index()
cur_row = {
    "Innings":["First","Second"],
    "Runs" : [Innings_score.iloc[0,1] + Innings_score.iloc[2,1],
              Innings_score.iloc[1,1] + Innings_score.iloc[3,1]]
   }
df = pd.DataFrame(cur_row)
df
```

```python
#top 3 batsmen
batsman_score = ball[['Striker_Id','Batsman_Scored']].groupby('Striker_Id').sum().reset_index()
batsman_score = pd.merge(player,batsman_score,how='left',left_on='Player_Id',right_on='Striker_Id')
batsman_score[['Player_Name','Batsman_Scored']].sort_values(by='Batsman_Scored',ascending=False).head(3)
```

```python
#top 10 bowler based on extras
bowler_extras = ball[ball['Extra_Runs']!=' '].groupby('Bowler_Id')['Extra_Runs'].sum().reset_index()
top_bowler_extras = pd.merge(player,bowler_extras,how='left',left_on='Player_Id',right_on='Bowler_Id')
top_bowler_extras[['Player_Name','Extra_Runs']].sort_values(by='Extra_Runs',ascending=False).head(10)
```

```python
#Top3 batsman based on centuries
batsman_score_match = ball.groupby(['Striker_Id','Match_Id'])['Batsman_Scored'].sum().reset_index()
batsman_score_centuries = batsman_score_match[batsman_score_match['Batsman_Scored'] >  99 ]
batsman_score_centuries = batsman_score_centuries.groupby('Striker_Id')['Batsman_Scored'].count().reset_index()
highest_centuries = pd.merge(player,batsman_score_centuries,how='left',left_on='Player_Id',right_on='Striker_Id')
highest_centuries[['Player_Name','Batsman_Scored']].sort_values(by = 'Batsman_Scored',ascending=False).head(3)
```

```python
#Are right handed batsman dominating IPL?
runs_player = ball.groupby('Striker_Id')['Batsman_Scored'].sum().reset_index()
highest_runs = pd.merge(player,runs_player,how='left',left_on='Player_Id',right_on='Striker_Id')
Batting_hand = highest_runs.groupby('Batting_Hand')['Batsman_Scored'].sum().reset_index()
player_hand = player.groupby('Batting_Hand')['Player_Id'].count().reset_index()
Batting_hand.insert(1,'No of players',player_hand['Player_Id'])
Batting_hand['Average Score'] = Batting_hand['Batsman_Scored'] / Batting_hand['No of players'] 
Batting_hand
```
