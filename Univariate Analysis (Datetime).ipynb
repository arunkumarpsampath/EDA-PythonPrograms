```python
moditweets = pd.read_excel('narendramodi_tweets.xlsx')
moditweets.head()
```

```python
moditweets['month'] = moditweets['created_at'].dt.month
moditweets['hour'] = moditweets['created_at'].dt.hour
moditweets['day'] = moditweets['created_at'].dt.day_name()
moditweets.head()
```

```python
#month wise tweets

moditweets.groupby('month')['id'].count().reset_index()
```

```python
#hour wise tweets

moditweets.groupby('hour')['id'].count().reset_index().sort_values(by='id',ascending=False)
```

```python
#source wise - week-day wise tweets
moditweets.groupby(['source','day'])['id'].count().reset_index().sort_values(by='id',ascending=False)
```
