# High score and low scoring matches

```python
matchscore = ball.groupby(['Season_Id','Match_Id'])['Batsman_Scored'].sum().reset_index()
highscorematches = matchscore[matchscore['Batsman_Scored'] > 300].groupby('Season_Id').count()
highscorematches = highscorematches.rename(columns = {'Match_Id': 'Score > 300'})
highscorematches = highscorematches[['Score > 300']]
highscorematches = highscorematches.reset_index()
highscorematches
```

```python
lowscorematches = matchscore[matchscore['Batsman_Scored'] <= 300].groupby('Season_Id').count()
lowscorematches = lowscorematches.rename(columns = {'Match_Id': 'Score <= 300'})
lowscorematches = lowscorematches[['Score <= 300']]
lowscorematches = lowscorematches.reset_index()
lowscorematches
```

```python
highlowscorematch = lowscorematches.merge(highscorematches)
highlowscorematch
```

```python
highlowscorematch['% low'] = highlowscorematch['Score <= 300'] / (highlowscorematch['Score <= 300'] + highlowscorematch['Score > 300'])
highlowscorematch['% high'] = 1 - highlowscorematch['% low']
highlowscorematch
```
