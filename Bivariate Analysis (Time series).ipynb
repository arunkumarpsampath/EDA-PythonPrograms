```python
import os
import pandas as pd
from matplotlib import pyplot as plt
os.chdir("D:\Manipal Pro learn\Term1\EDA\Dataset")
```

```python
dailytemp = pd.read_excel("daily-minimum-temperatures.xlsx")
dailytemp.head()
```

```python
#Line chart analysis

plt.plot('Date','Daily minimum temperatures',data=dailytemp, color = 'B')
#to plot another line
#plt.plot('Date','Humidity',data=x,color='R')
```

```python
from datetime import date
```

```python
dailytemp['month'] = dailytemp['Date'].dt.month_name()
dailytemp['year'] = dailytemp['Date'].dt.year
dailytemp.head()
```

```python
new_temp = dailytemp.groupby(['year','month'])['Daily minimum temperatures'].mean().reset_index()
new_temp.head()
```

```python
new_temp_wide = new_temp.pivot('month','year','Daily minimum temperatures')
new_temp_wide
```

```python
index_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
new_temp_wide = new_temp_wide.reindex(index_order)
new_temp_wide
```

```python
import seaborn as sns
```

```python
sns.heatmap(new_temp_wide, cmap='RdYlGn_r',linewidths=0.5)
```

```python
dailytemp['day'] = dailytemp['Date'].dt.day
dailytemp.head()
```

```python
new_temp = dailytemp.groupby(['month','day'])['Daily minimum temperatures'].mean().reset_index()
new_temp.head()
```

```python
day_mon_temp = new_temp.pivot('day','month','Daily minimum temperatures')
day_mon_temp.head()
```

```python
day_mon_temp = day_mon_temp[index_order]
sns.heatmap(day_mon_temp, cmap='RdYlGn_r',linewidths=0.5,annot=True)
```
