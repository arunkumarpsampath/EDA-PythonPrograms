# KNN imputation

```python
from missingpy import KNNImputer
```

```python
missing = pd.read_excel("Missingvalues.xlsx")
missing.head()
```

```python
# creating the imputation object
imputer = KNNImputer(missing_values='NaN',metric = 'masked_euclidean',n_neighbors=2,weights='distance')
#missing_values - list of represention of missing values in given data
#metric - which type of distance calculation to be used
#n_neighbours - # of nearest neighbours to be taken to calculate the missing values
#ex: n_neighbours=2 it will get average distance bwn two nearest points
#weights='distance' - gives weight based on inverse of distance
#weights='uniform' - gives equal weight for all neighbours
```

```python
#for single column it replaces with average of that column
new_demo = imputer.fit_transform(missing[['X3']])
new_demo
```

```python
(23+45+56+56)/4
```

```python
#for whole df it replaces with average of nearest neighbours as specified
new_demo = imputer.fit_transform(missing)
new_demo
```

# Kmeans Clustering

```python
from pydataset import data
```

```python
df = data('iris')
df.head()
```

```python
#removing last column
df1 = df.drop(columns=['Species'])
df1.head()
```

```python
from sklearn.cluster import KMeans
```

```python
#create kmeans object
km = KMeans(n_clusters=3,init='k-means++',random_state=1234)
```

```python
#fit the model on our data
km.fit(df1.values)
```

```python
#Identifying the centroids co-ordinate

km.cluster_centers_
```

```python
#creating the column in our dataset
df1['Cluster'] = km.predict(df1.values)
df1.head()
```

```python
df1.groupby('Cluster').size()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(df1['Sepal.Length'],df1['Sepal.Width'],c = df1['Cluster'])
plt.xlabel('Sepal length')
plt.ylabel("Sepal width")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='red',s=100)
```

```python
plt.scatter(df1['Petal.Length'],df1['Petal.Width'],c = df1['Cluster'])
plt.xlabel('Petal length')
plt.ylabel("Petal width")

plt.scatter(km.cluster_centers_[:,2],km.cluster_centers_[:,3],c='red',s=100)
```

```python
election = pd.read_excel("Election data base.xlsx")
election.head()
```

```python
statewisevotelectors = election.groupby('st_name')['totvotpoll','electors'].mean().reset_index()
statewisevotelectors.head()
```

```python
km.fit(statewisevotelectors[['totvotpoll','electors']])
```

```python
km.cluster_centers_
```

```python
km.groupby('Cluster').size()
```

```python
statewisevotelectors['Cluster'] = km.predict(statewisevotelectors[['totvotpoll','electors']])
statewisevotelectors.head()
```

```python
plt.scatter(statewisevotelectors['electors'],statewisevotelectors['totvotpoll'],c = statewisevotelectors['Cluster'])
plt.xlabel('Electors')
plt.ylabel("Total polled")

plt.scatter(km.cluster_centers_[:,1],km.cluster_centers_[:,0],c='red',s=100)
```

# Elbow Curve to determine optimal cluster size

```python
km1= KMeans(n_clusters=1,init='k-means++')
```

```python
km1.fit(df1.values)
```

```python
km1.inertia_    #gives error
```

```python
km2= KMeans(n_clusters=2,init='k-means++')
```

```python
km2.fit(df1.values)
```

```python
km2.inertia_
```

```python
cluster_error = []

for i in range(1,15):
    kmeans = KMeans(n_clusters= i , init = 'k-means++',random_state=1234)
    kmeans.fit(df1.values)
    cluster_error.append(kmeans.inertia_)
cluster_error
```

```python
plt.plot(range(1,15),cluster_error,marker='o')
```

```python
election_error = []

for i in range(1,15):
    km = KMeans(n_clusters=i,init='k-means++',random_state=1234)
    km.fit(statewisevotelectors[['totvotpoll','electors']])
    election_error.append(km.inertia_)

election_error
```

```python
plt.plot(range(1,15),election_error,marker='.')
```
