```python
#find top20 products - Creating association for whole data takes long time

prodcount = order_product.groupby('product_id').size().reset_index()

prodcount = prodcount.rename(columns={0:'Freq'})

top20products= prodcount.sort_values(by='Freq',ascending=False).head(20)
top20products
```

```python
top20productsname = pd.merge(top20products,products)
top20productsname.head()
```

```python
top20 = top20productsname[['product_id','product_name']]
top20
```

```python
orderstop20 = order_product[order_product['product_id'].isin(top20['product_id'])]
orderstop20
```

```python
len(orderstop20['order_id'].unique())
```

```python
#lets create the long table of order_id and product_name
ordertop20prodname = pd.merge(orderstop20[['order_id','product_id']],products)
ordertop20prodname.head()
```

```python
ordertop20prod = ordertop20prodname.groupby(['order_id','product_name']).size().reset_index()
ordertop20prod.head()
```

```python
ordertop20prod = ordertop20prod.rename(columns={0:'count'})
ordertop20prod.head()
```

```python
orderpivot = ordertop20prod.pivot('order_id','product_name','count')
orderpivot
```

```python
orderpivot = orderpivot.fillna(0)
orderpivot.head()
```

```python
#creating frquently occuring itemset

freq_itemsets = apriori(orderpivot, min_support=0.01,use_colnames=True)
```

```python
freq_itemsets.head()
```

```python
freq_itemsets.sort_values(by='support',ascending=False).head(25)
```

```python
#creating rules for available set
rules = association_rules(freq_itemsets,metric='lift',min_threshold=1.001)
```

```python
rules.head()
#antecedents - X
#consequents - Y
#support  - support(x,y)
#confidence - p(y/x)
```

```python
rules.sort_values(by='lift',ascending=False).head()
```
