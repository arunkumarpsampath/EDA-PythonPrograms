```python
df = pd.DataFrame({'age':[2,4,6],
                   'height' : [20,40,60],
                  'weight' : [32,34,38],
                  'gender' : ['M','F','M']})
df.head()
```

```python
x = df[['age','height','weight']]

#calculating eigen values and eigen vector

evalue,evector = np.linalg.eig(x)
```

```python
evalue
```

```python
evector
```

```python
pcs = np.dot(x,evector)
pcs
```

```python
pcs_df=pd.DataFrame(pcs)
pcs_df
```

```python
# calculate amount of information from each pc
var = pcs_df.var()
var = pd.DataFrame(var)
var
```

```python
var['%'] = round((var[0] /pcs_df.var().sum() * 100),2)
var = var.rename(columns={0:'Variance of pcs'})
var
```

```python
from matplotlib import pyplot as plt
x_axis=[1,2,3]
y_axis = var['%']
plt.plot(x_axis,y_axis)
```

```python
from pydataset import data

iris = data('iris')
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
x= iris[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']]
x
```

```python
y=iris[['Species']]
y.head()
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#splitting data
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2)
```

```python
print(len(X_train),len(X_test),len(Y_train),len(Y_test))
```

```python
#standardize your data
from sklearn.preprocessing import StandardScaler

sc1 = StandardScaler()
X_train1 = sc1.fit_transform(X_train)
X_test1 = sc1.fit_transform(X_test)
```

```python
X_train1.mean()
```

# using Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier

clasifier = RandomForestClassifier()

rfmodel = clasifier.fit(X_train1,Y_train)
rfmodel
```

```python
#predicting test data

Y_test['Predicted'] = rfmodel.predict(X_test1)
```

```python
wronglypred = Y_test[~(Y_test['Species']==Y_test['Predicted'])]
wronglypred
```

```python
pd.crosstab(Y_test['Species'],Y_test['Predicted'])
```
